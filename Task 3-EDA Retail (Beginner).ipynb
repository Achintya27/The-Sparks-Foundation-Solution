#### Correlation between every int columns

```python
df.corr()
```

```python
fig,axes = plt.subplots(1,1,figsize=(10,6))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

#### Covariance between every int columns

```python
df.cov()
```

```python
fig,axes = plt.subplots(1,1,figsize=(10,6))
sns.heatmap(df.cov(), annot= True)
plt.show()
```

### Counts of unique values in every columns

```python
for i in df.columns:
    x = df[i].value_counts()
    print("Column name is:",i,"and it value is:",x)
    print()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python

plt.figure(figsize=(20,12))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x=df['Sub-Category'], y=df['Profit'])
plt.xticks(rotation=90)
plt.show()
```

```python
df.hist(bins=50 ,figsize=(20,15))
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = df, color = 'r', label= 'Discount')
plt.legend()
```

```python
plt.figure(figsize = (100,20))
sns.lineplot('Sales', 'Profit', data = df, color = 'r', label= 'Sales')
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.legend()
```

```python
plt.figure(figsize=(20,10))
state1 = df.groupby(['State'])['Profit'].mean().reset_index()
sns.barplot(x = 'State', y='Profit', data=state1)

plt.xticks(rotation=90)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('State', fontsize=20) 
plt.ylabel('Mean Profit', fontsize=20)
plt.show()
```

```python
plt.style.use('fivethirtyeight')

plt.pie(x= df['Country'], labels=df['Country'], wedgeprops = {'edgecolor':'black'})

plt.title("My Awsome Pie Chart")
plt.tight_layout()
plt.show()
```
